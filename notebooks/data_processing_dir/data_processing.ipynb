```python
#ingredient_targets data prossing
import pandas as pd

# 读取成分靶点的三个数据表
df1_1 = pd.read_csv('source1_table1.csv')
df1_2 = pd.read_csv('source1_table2.csv')
df1_3 = pd.read_csv('source1_table3.csv')

# 提取A列并去重
drug_target = pd.concat([df1_1['A'], df1_2['A'], df1_3['A']]).drop_duplicates().reset_index(drop=True)

# 读取疾病靶点的三个数据表
df2_1 = pd.read_csv('source2_table1.csv')
df2_2 = pd.read_csv('source2_table2.csv')
df2_3 = pd.read_csv('source2_table3.csv')

# 提取A列并去重
disease_target = pd.concat([df2_1['A'], df2_2['A'], df2_3['A']]).drop_duplicates().reset_index(drop=True)

# 生成新的数据表
new_df = pd.DataFrame({
    'drug_target': drug_target,
    'disease_target': disease_target
})

# 保存新的数据表为CSV文件
new_df.to_csv('new_table.csv', index=False)
```

```python
import pandas as pd

def merge_and_deduplicate_columns(source_files, target_column_name):
    """
    从多个数据源文件中提取指定列，合并并去重，返回一个Series。

    参数:
    source_files (list of str): 数据源文件的路径列表。
    target_column_name (str): 需要提取的列名。

    返回:
    pd.Series: 合并并去重后的列数据。
    """
    # 初始化一个空列表用于存储每个数据表的目标列
    columns = []

    # 遍历所有数据源文件并提取目标列
    for file in source_files:
        df = pd.read_csv(file)
        columns.append(df[target_column_name])

    # 合并所有列并去重
    merged_column = pd.concat(columns).drop_duplicates().reset_index(drop=True)
    return merged_column

def create_new_table(drug_files, disease_files, drug_column, disease_column):
    """
    创建一个新的数据表，包含去重后的drug_target和disease_target列。

    参数:
    drug_files (list of str): 包含药物目标数据的文件列表。
    disease_files (list of str): 包含疾病目标数据的文件列表。
    drug_column (str): 药物目标数据的列名。
    disease_column (str): 疾病目标数据的列名。

    返回:
    pd.DataFrame: 新的数据表。
    """
    drug_target = merge_and_deduplicate_columns(drug_files, drug_column)
    disease_target = merge_and_deduplicate_columns(disease_files, disease_column)

    new_df = pd.DataFrame({
        'drug_target': drug_target,
        'disease_target': disease_target
    })

    return new_df

# 示例文件列表
drug_files = ['source1_table1.csv', 'source1_table2.csv', 'source1_table3.csv']
disease_files = ['source2_table1.csv', 'source2_table2.csv', 'source2_table3.csv']

# 创建新的数据表
new_table = create_new_table(drug_files, disease_files, 'A', 'A')

# 保存新的数据表为CSV文件
new_table.to_csv('new_table.csv', index=False)

print(new_table)
```
